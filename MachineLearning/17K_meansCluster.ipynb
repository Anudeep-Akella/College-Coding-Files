```python
import pandas as  pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

df = pd.read_csv('/content/call_center_kmeans_50 (1).csv')
df.head()
```

```python
loses = []
for i in range(1,10):
  kms = KMeans(n_clusters=i)
  kms.fit(df)

  loses.append(kms.inertia_)

print(loses)
```

```python
plt.plot(range(1,10),loses)
plt.show()
```
