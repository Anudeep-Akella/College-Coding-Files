```python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Step 1: Create a non-linear dataset
np.random.seed(42)
X = np.sort(5 * np.random.rand(100, 1), axis=0)
y = 2 + 1.5 * X**2 + np.random.randn(100, 1)

# Step 2: Convert to Polynomial Features (degree = 2)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Step 3: Train the Polynomial Regression Model
model = LinearRegression()
model.fit(X_poly, y)

# Step 4: Make Predictions
X_range = np.linspace(0, 5, 100).reshape(-1, 1)
X_range_poly = poly.transform(X_range)
y_pred = model.predict(X_range_poly)

# Step 5: Evaluation Metrics
print("Mean Squared Error (MSE):", mean_squared_error(y, model.predict(X_poly)))
print("R2 Score:", r2_score(y, model.predict(X_poly)))

# Step 6: Visualization
plt.scatter(X, y, color='blue', label="Original Data")
plt.plot(X_range, y_pred, color='red', linewidth=2, label="Polynomial Regression Curve")
plt.xlabel("X")
plt.ylabel("y")
plt.title("Polynomial Regression (Degree 2)")
plt.legend()
plt.show()
```
