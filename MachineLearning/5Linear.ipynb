Give the .csv file to the data frames. Here in this linear regression we are using the advertising data set

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1. Load the data
# Ensure the advertising.csv file is in your local directory or Colab path
df = pd.read_csv('/content/advertising.csv')

# Defining features (x) and target (y)
x = df['TV'].values
y = df['Sales'].values

# 2. Initialize parameters
w0 = 0.0  # Intercept
w1 = 0.0  # Slope
lr = 0.00001  # Learning rate (small for stability)
epochs = 5000 # Number of iterations for convergence

# 3. Gradient Descent Loop
for i in range(epochs):
    yh = w0 + w1 * x

    # Calculate Gradients
    dew0 = -2 * (y - yh).mean()
    dew1 = -2 * ((y - yh) * x).mean()

    # Update weights
    w0 = w0 - lr * dew0
    w1 = w1 - lr * dew1

# 4. Final Predictions
yh = w0 + w1 * x

# 5. Calculate Metrics
mse = ((y - yh)**2).mean()
mae = (np.abs(y - yh)).mean()

# R-squared calculation
ss_res = ((y - yh)**2).sum()      # Sum of Squares Residuals
ss_tot = ((y - y.mean())**2).sum() # Total Sum of Squares
r2 = 1 - (ss_res / ss_tot)

print("--- Final Evaluation Metrics ---")
print(f"Mean Squared Error (MSE): {mse:.4f}")
print(f"Mean Absolute Error (MAE): {mae:.4f}")
print(f"R-squared (R2): {r2:.4f}")
print(f"Final Equation: Sales = {w0:.4f} + {w1:.4f} * TV")

# 6. Visualization
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='blue', label='Actual Data', alpha=0.5)
plt.plot(x, yh, color='red', linewidth=2, label='Regression Line')
plt.title('TV Advertising vs Sales')
plt.xlabel('TV Budget')
plt.ylabel('Sales')
plt.legend()
plt.show()
```
