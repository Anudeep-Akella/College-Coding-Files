```python
import numpy as np

# Activation Function: Sigmoid
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Derivative of Sigmoid for Backpropagation
def sigmoid_derivative(x):
    return x * (1 - x)

# 1. Dataset (XOR Logic Gate)
X = np.array([[0,0], [0,1], [1,0], [1,1]])
y = np.array([[0], [1], [1], [0]])

# 2. Hyperparameters
input_layer_neurons = 2
hidden_layer_neurons = 3
output_neurons = 1
learning_rate = 0.5
epochs = 10000

# 3. Weight and Bias Initialization
# Weights connect layers: Input -> Hidden -> Output
wh = np.random.uniform(size=(input_layer_neurons, hidden_layer_neurons))
bh = np.random.uniform(size=(1, hidden_layer_neurons))
wout = np.random.uniform(size=(hidden_layer_neurons, output_neurons))
bout = np.random.uniform(size=(1, output_neurons))

# 4. Training Loop
for i in range(epochs):
    # --- FORWARD PROPAGATION ---
    hidden_layer_input = np.dot(X, wh) + bh
    hidden_layer_activations = sigmoid(hidden_layer_input)

    output_layer_input = np.dot(hidden_layer_activations, wout) + bout
    predicted_output = sigmoid(output_layer_input)

    # --- BACKPROPAGATION ---
    # Step A: Calculate Error at Output
    error = y - predicted_output
    d_predicted_output = error * sigmoid_derivative(predicted_output)

    # Step B: Calculate Error at Hidden Layer
    error_hidden_layer = d_predicted_output.dot(wout.T)
    d_hidden_layer = error_hidden_layer * sigmoid_derivative(hidden_layer_activations)

    # --- WEIGHT UPDATES (Gradient Descent) ---
    wout += hidden_layer_activations.T.dot(d_predicted_output) * learning_rate
    bout += np.sum(d_predicted_output, axis=0, keepdims=True) * learning_rate
    wh += X.T.dot(d_hidden_layer) * learning_rate
    bh += np.sum(d_hidden_layer, axis=0, keepdims=True) * learning_rate

    if i % 2000 == 0:
        loss = np.mean(np.square(y - predicted_output))
        print(f"Epoch {i} - Loss: {loss:.4f}")

print("\nFinal Predicted Output After Training:")
print(predicted_output)
```
