```python
import pandas as pd

class RuleBasedClassifier:
    def __init__(self):
        self.rules = []

    def add_rule(self, condition_func, label):
        """Adds a custom rule: a function that returns True/False and a resulting label."""
        self.rules.append((condition_func, label))

    def classify(self, data_point):
        """Check the data against all rules and return the first match."""
        for condition, label in self.rules:
            if condition(data_point):
                return label
        return "Unknown"  # Default if no rules match

# --- 1. Define the Data ---
data = [
    {"name": "Campaign A", "spend": 500, "clicks": 50},
    {"name": "Campaign B", "spend": 2000, "clicks": 10},
    {"name": "Campaign C", "spend": 50, "clicks": 5},
    {"name": "Campaign D", "spend": 1200, "clicks": 300},
]

# --- 2. Define Custom Rules ---
# Each rule is a lambda or function that takes a data point and returns a boolean
def is_high_performer(row):
    return row['clicks'] > 100 and row['spend'] < 1500

def is_money_pit(row):
    return row['spend'] > 1000 and row['clicks'] < 20

def is_low_budget(row):
    return row['spend'] < 100

# --- 3. Initialize and Run ---
classifier = RuleBasedClassifier()

# Order matters: rules are evaluated top to bottom
classifier.add_rule(is_high_performer, "High Performance")
classifier.add_rule(is_money_pit, "Inefficient")
classifier.add_rule(is_low_budget, "Small Scale")
classifier.add_rule(lambda x: x['spend'] >= 100, "Standard")

# Apply to data
results = []
for entry in data:
    category = classifier.classify(entry)
    entry['category'] = category
    results.append(entry)

# Display results
df_results = pd.DataFrame(results)
print(df_results)
```
